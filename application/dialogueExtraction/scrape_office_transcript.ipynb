# The Office transcripts, process, and save to CSV file

From: http://officequotes.net/no1-01.php

```python
from bs4 import BeautifulSoup
import requests
import os
import re
```

```python
# General stuff
officeDir = "../../contentData/tvShows/The-Office/"
```

```python
linkHome = "http://officequotes.net/"
linkMain = "http://officequotes.net/no1-01.php"
page = requests.get(linkMain) # try the following

soup = BeautifulSoup(page.content, 'html5lib')
```

```python
def remove_stage_directions(data):
    """remove things between square brackets [...]
    """
    p = re.compile(r'\[.*?\]')
    return p.sub('', data)


def remove_parens(data):
    """remove things between parens (...)
    """
    p = re.compile(r'\(.*?\)')
    return p.sub('', data)
```

```python
def getOfficeDialogueDichotomy(linkToEpisode):
    page = requests.get(linkToEpisode)
    soup = BeautifulSoup(page.content, 'html5lib')
    quotes = soup.find_all('div', class_='quote')
    dd = []
    PUNCTUATION = '!"#$%&()*+,-./;<=>?@[\\]^_`{|}~'
    for quote in quotes:
        txt = quote.text
        txt = txt.replace("\t", "").strip()
        if ("Deleted Scene" in txt): break
        txt = txt.split("\n")
        for diag in txt:
            # CONTINUE NOTE: check if this extracts charNames and dialogues
            tmp = diag.split(":")
            if (len(tmp) < 2): continue # skip
            charName = tmp[0].strip().upper()
            charName = charName.translate(str.maketrans(PUNCTUATION, ' '*len(PUNCTUATION)))
            dialogue = " ".join(tmp[1:])
            dialogue = remove_stage_directions(dialogue)
            dialogue = remove_parens(dialogue)
            dialogue = dialogue.strip().lower().translate(str.maketrans(PUNCTUATION, ' '*len(PUNCTUATION)))
            dialogue = " ".join(dialogue.split())
            if (charName is "" or dialogue is ""): continue # skip
            dd.append((charName, dialogue))
    return dd
```

```python
transcripts = getOfficeDialogueDichotomy(linkMain)
```

```python
def writeToCSV(transcriptsDichotomy, fileName, delim=';'):
    with open(fileName, 'w') as file:
        for diag in transcriptsDichotomy:
            file.write(diag[0] + delim + diag[1] + "\n")
        print("Done writing to CSV")
```

```python
allEpisodes = os.path.join(officeDir, "metaAllEpisodes.txt")
with open(allEpisodes, 'r') as file:
    lines = file.readlines()
    for line in lines:
        tmp = line.replace("\n", "").split(";")
        episodeDir = os.path.join(officeDir, str(tmp[0]))
        episodeDir = os.path.join(episodeDir, str(tmp[1])+"-"+tmp[2])
        print("dir:", episodeDir)
        link = tmp[3]
        print("link:", link)
        transcripts = getOfficeDialogueDichotomy(link)
        csvName = os.path.join(episodeDir, "csvTranscript.csv")
        writeToCSV(transcripts, csvName)
```

# Get Links to The Office transcripts

From: http://officequotes.net/no1-01.php

```python
from bs4 import BeautifulSoup
import requests

# import re

```

```python
# General stuff
officeDir = "../../contentData/tvShows/The-Office/"
```

```python
linkHome = "http://officequotes.net/"
linkMain = "http://officequotes.net/no1-01.php"
page = requests.get(linkMain) # try the following

soup = BeautifulSoup(page.content, 'html5lib')
```

```python
contents = soup.find_all('a')
contents = contents[8:-9]
print(contents[0])
print(contents[-1])
```

```python
ep = contents[0]
ep = str(ep)
ep
```

```python
import re
```

```python
print(re.findall(r'(?:href=\"no)(.*?)(?=.php\")', ep))
print(re.findall(r'(?:href=\")(.*?)(?=\")', ep))
```

```python
import os
```

```python
PUNCTUATION = '!"#$%&()\'*+,-./;<=>?@[\\]^_`{|}~'
allMetaStrings = []
for ep in contents:
    title = ep.text
#     title = title.translate(str.maketrans(PUNCTUATION, ''*len(PUNCTUATION)))
    title = title.translate(str.maketrans('', '', PUNCTUATION))
    title = title.replace(' ', '-')
#     print("title:\t", title)
    ep = str(ep)
    link = re.findall(r'(?:href=\")(.*?)(?=\")', ep)[0]
    link = os.path.join(linkHome, link)
#     print("link:\t", link)
    tmp = re.findall(r'(?:href=\"no)(.*?)(?=.php\")', ep)[0].split('-')
    season = int(tmp[0])
    episode = int(tmp[1])
#     print("season:\t", season)
#     print("episode:", episode)
    episodeFolderName = "-".join([str(episode), title])
    episodeFolderPath = os.path.join(officeDir, str(season))
    episodeFolderPath = os.path.join(episodeFolderPath, episodeFolderName)
#     print("folder:\t", episodeFolderPath)
    os.makedirs(episodeFolderPath)
    metaFile = os.path.join(episodeFolderPath, "metaData.txt")
    metaString = ";".join([str(season), str(episode), title, link])
    allMetaStrings.append(metaString)
    print(metaString)
    with open(metaFile, 'w') as file:
        file.write(metaString + '\n')
    
```

```python
allMetaStrings
```

```python
# write all episodes to a file in home folder
allMetaFile = os.path.join(officeDir, "metaAllEpisodes.txt")
with open(allMetaFile, 'w') as file:
    for m in allMetaStrings:
        file.write(m + "\n")
```
